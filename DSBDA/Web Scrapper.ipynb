```python
import requests
import bs4
```

```python
req1 = requests.get('https://dl.flipkart.com/s/MTkhQ0NNNN')
req1
```

```python
req1.content
```

```python
soup = bs4.BeautifulSoup(req1.text)
soup
```

```python
reviews = soup.findAll('div',{'class' : 'RcXBOT'});
for review in reviews:
    print(review.get_text())
```

```python
ratings = soup.find('div',{'class' : 'ggs1+C'}).get_text();
print(ratings)
```

```python
individual_rating = soup.findAll('div', {'class' : 'Ga3i8K'});
for ind_rating in individual_rating:
    print(ind_rating.get_text())
```

```python
tags = soup.find('span', {'class' : '_6EBuvT OHdLLG'}).get_text();
tags
```

```python
customer_name = soup.findAll('p',{'class' : '_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.get_text())
```
