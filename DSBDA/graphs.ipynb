```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
df=pd.read_csv('forestfires.csv')
df
```

```python

plt.scatter(df['month'], df['wind'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Month')
plt.ylabel('wind')

 
plt.show()
```

```python
plt.bar(df['day'], df['temp'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Temprature')
 
# Adding the legends
plt.show()
```

```python
plt.hist(df['DMC'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```

```python
keys='FFMC','DMC','DC'
sizes=[df['FFMC'].sum(),df['DMC'].sum(),df['DC'].sum()]
# Plotting data on chart



plt.pie(sizes, labels=keys, autopct='%.0f%%')

# Add title to the chart
plt.title('Values of fire weather indices')

# Displaying chart
plt.show()
```

```python
df
df.boxplot(by ='day', column =['temp'], grid = False)
```

```python
sn.scatterplot(x='month', y='temp', data=df,hue='day')
plt.title('Scatter Chart')
plt.show()
```

```python
sn.lineplot(x="month", y="temp", data=df)
 
# setting the title 
plt.title('Line Chart')
 
plt.show()
```

```python
sn.set_style("whitegrid")
  
sn.boxplot(x = 'day', y = 'temp', data = df)
```

```python
data = np.random.rand(10, 12)  # 10 rows, 12 columns
df = pd.DataFrame(data, columns=[f'Month {i+1}' for i in range(12)])

# Step 3: Create a heatmap
plt.figure(figsize=(12, 8))
sn.heatmap(df, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=0.5)
plt.title('Sample Heatmap')
plt.xlabel('Months')
plt.ylabel('Rows')
plt.show()
```
